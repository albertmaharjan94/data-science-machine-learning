## Matplotlib example workflow

```python
# 0. import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fix, ax = plt.subplots(figsize=(20,20))

# 3. Plot data
ax.plot(x, y)
# 4. Customize plot 
ax.set(title = "Simple Plot",
      xlabel = "X-axis",
      ylabel = "Y-axis")

# 5. Save and whow figure(whole)
fig.savefig("images/test.png")
```

## Making figures with numpy arrays
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# create data 
x = np.linspace(0,10,100)
x[:10]
```

```python
# plot // default is lineplot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# scatter plot
fix, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# another scatter
fix, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# making plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 5,
                    "Cashew butter": 19}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),  nut_butter_prices.values())
ax.set(title="Nut butter store",
      xlabel="Butter",
      ylabel="Price($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), nut_butter_prices.values())
```

```python
nut_butter_prices.keys(), nut_butter_prices.values()
```

```python
# make some data for histogram and plot
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# subplot option 1
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(nrows = 2,
                                            ncols = 2,
                                            figsize =(10,5))
# Plot to each different axes
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplots option 2
fix, ax = plt.subplots(nrows = 2,
                        ncols = 2,
                        figsize = (10,5))

# plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```
